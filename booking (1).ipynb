# **Introduction**
Dataset Name: NCR Ride Bookings Dataset   (National Capital Region)

Description:

This dataset contains 150,000 ride booking records from a ride-hailing/transportation platform in the NCR region.

It captures detailed information about bookings, ride status, customer/driver behavior, trip metrics, ratings, cancellations, and payment methods.

# **Data** **Cleaning**

isnull

To identify columns with incomplete data.
it show the column null values count

```python
booking.isnull().sum()
```

total missing values

```python
booking.isnull().sum().sum()
```

fillna

To treat missing data by filling, removing, or keeping it appropriately.

```python
booking['Avg CTAT']=booking['Avg CTAT'].fillna(booking['Avg CTAT'].median())
```

```python
booking['Avg VTAT']=booking['Avg VTAT'].fillna(booking['Avg VTAT'].median())
```

```python
booking['Booking Value']=booking['Booking Value'].fillna(booking['Booking Value'].median())
```

```python
booking['Ride Distance']=booking['Ride Distance'].fillna(booking['Ride Distance'].median())
```

```python
booking['Driver Ratings']=booking['Driver Ratings'].fillna(3.9)
```

```python
booking['Customer Rating']=booking['Customer Rating'].fillna(5)
```

```python
booking['Payment Method']=booking['Payment Method'].fillna('Gpay')
```

**remove unwanted** **column**

```python
booking.drop([
    'Cancelled Rides by Customer',
    'Reason for cancelling by Customer',
    'Cancelled Rides by Driver',
    'Driver Cancellation Reason',
    'Incomplete Rides',
    'Incomplete Rides Reason'
], axis=1, inplace=True)
```

Now check null values

```python
booking.isnull().sum()
```

check column and data type

```python
booking.info()
```

dupicated

To eliminate repeated records.

```python
booking.duplicated().sum()
```

Fix data types

```python

booking['Date'] = pd.to_datetime(booking['Date'])

booking['Time'] = pd.to_datetime(booking['Time']).dt.time

booking['Booking Status'] = booking['Booking Status'].astype('category')
booking['Vehicle Type'] = booking['Vehicle Type'].astype('category')
booking['Pickup Location'] = booking['Pickup Location'].astype('category')
booking['Drop Location'] = booking['Drop Location'].astype('category')
booking['Payment Method'] = booking['Payment Method'].astype('category')

booking['Booking Value'] = booking['Booking Value'].astype('Int64')
```

IQR

```python
Q1 = booking['Booking Value'].quantile(0.25)
Q3 = booking['Booking Value'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

# **Univariate Analysis**

# **trend comparisons**

Month over month

```python

booking['Date'] = pd.to_datetime(booking['Date'])

booking['Month'] = booking['Date'].dt.to_period('M').astype(str)


mom_revenue = booking.groupby('Month')['Booking Value'].sum().reset_index()


fig = px.line(
    mom_revenue,
    x='Month',
    y='Booking Value',
    markers=True,
    title='Month-over-Month Revenue Trend'
)


fig.update_traces(line=dict(color='teal', width=3))
fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Total Revenue'
)

fig.show()
```

```python
booking['Date'] = pd.to_datetime(booking['Date'])


daily_revenue = booking.groupby('Date')['Booking Value'].sum().reset_index()


fig = px.line(
    daily_revenue,
    x='Date',
    y='Booking Value',
    markers=True,
    title='Daily Revenue Trend'
)


fig.update_traces(line=dict(color='teal', width=3))
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Total Revenue'
)

fig.show()
```

# **TREND ANALYSIS**

Upward Trend

```python
days = pd.date_range(start='2026-01-01', periods=100)
upward = pd.DataFrame({'Date': days, 'Value': np.arange(100)})

fig = px.line(upward, x='Date', y='Value', title='Upward Trend', line_shape='linear',color_discrete_sequence=['black'])
fig.show()
```

Downward Trend

```python
downward = pd.DataFrame({'Date': days, 'Value': np.arange(100,0,-1)})

fig = px.line(downward, x='Date', y='Value', title='Downward Trend', line_shape='linear',color_discrete_sequence=['green'])
fig.show()
```

Horizontal Trend

```python
horizontal = pd.DataFrame({'Date': days, 'Value': np.full(100, 50)})

fig = px.line(horizontal, x='Date', y='Value', title='Horizontal Trend', line_shape='linear',color_discrete_sequence=['orange'])
fig.show()
```

Cyclical Trend

```python
cyclical = pd.DataFrame({'Date': days, 'Value': 50 + 20*np.sin(np.linspace(0, 10*np.pi, 100))})

fig = px.line(cyclical, x='Date', y='Value', title='Cyclical Trend',color_discrete_sequence=['teal'])
fig.show()
```

Exponential Trend

```python
exponential = pd.DataFrame({'Date': days, 'Value': np.exp(np.linspace(0, 5, 100))})

fig = px.line(exponential, x='Date', y='Value', title='Exponential Trend',color_discrete_sequence=['red'])
fig.show()
```

Declining to Zero Trend

```python
decline_zero = pd.DataFrame({'Date': days, 'Value': np.linspace(100, 0, 100)})

fig = px.line(decline_zero, x='Date', y='Value', title='Declining to Zero Trend',color_discrete_sequence=['purple'])
fig.show()
```

V Shape Trend

```python
vshape = pd.DataFrame({'Date': days, 'Value': np.abs(np.linspace(-50, 50, 100))})

fig = px.line(vshape, x='Date', y='Value', title='V Shape Trend',color_discrete_sequence=['pink'])
fig.show()
```

Inverted V Trend

```python
inv_v = pd.DataFrame({'Date': days, 'Value': -np.abs(np.linspace(-50, 50, 100)) + 50})

fig = px.line(inv_v, x='Date', y='Value', title='Inverted V Trend',color_discrete_sequence=['blue'])
fig.show()
```

Seasonal Trend

```python
seasonal = pd.DataFrame({'Date': days, 'Value': 50 + 20*np.sin(np.linspace(0, 4*np.pi, 100))})

fig = px.line(seasonal, x='Date', y='Value', title='Seasonal Trend',color_discrete_sequence=['magenta'])
fig.show()
```

Noise Chart


```python
noise = pd.DataFrame({'Date': days, 'Value': np.random.normal(50, 5, 100)})

fig = px.scatter(noise, x='Date', y='Value', title='Noise Chart',color_discrete_sequence=['olive'])
fig.show()
```
